```python
class Book:
    def __init__(self, title, author, isbn):
        self.title = title
        self.author = author
        self.isbn = isbn
        self.is_issued = False

    def mark_as_issued(self):
        if not self.is_issued:
            self.is_issued = True
            return True
        return False

    def mark_as_returned(self):
        if self.is_issued:
            self.is_issued = False
            return True
        return False

    def display_details(self):
        status = "Issued" if self.is_issued else "Available"
        return f"Title: {self.title}, Author: {self.author}, ISBN: {self.isbn}, Status: {status}"
class Library:
    def __init__(self):
        self.books = []

    def add_book(self, title, author, isbn):
        new_book = Book(title, author, isbn)
        self.books.append(new_book)
        print(f"Book '{title}' added successfully.")

    def issue_book(self, isbn):
        book = self._find_book(isbn)
        if book and not book.is_issued:
            if book.mark_as_issued():
                print(f"Book '{book.title}' issued successfully.")
            else:
                print("Book is already issued.")
        else:
            print("Book not found or already issued.")

    def return_book(self, isbn):
        book = self._find_book(isbn)
        if book and book.is_issued:
            if book.mark_as_returned():
                print(f"Book '{book.title}' returned successfully.")
            else:
                print("Book was not issued.")
        else:
            print("Book not found or was not issued.")

    def check_availability(self, isbn):
        book = self._find_book(isbn)
        if book:
            print(book.display_details())
        else:
            print("Book not found.")

    def _find_book(self, isbn):
        for book in self.books:
            if book.isbn == isbn:
                return book
        return None
def main():
    library = Library()
    # Adding books
    library.add_book("stochastic processes", "j medhi", "1234")
    library.add_book("xyz", "zbc", "5678")

    # Check availability of books
    library.check_availability("1234")
    library.check_availability("5678")

    # Issue a book
    library.issue_book("1234")

    # Check availability after issuing
    library.check_availability("1234")

    # Return the book
    library.return_book("12345")

    # Check availability after returning
    library.check_availability("1234")

    # Attempt to issue a non-existent book
    library.issue_book("012")

if __name__ == "__main__":
    main()
```

```python
import math

class Fraction:
    def __init__(self, numerator, denominator):
        if denominator == 0:
            raise ValueError("Denominator cannot be zero.")
        self.numerator = numerator
        self.denominator = denominator
        self.simplify()  # Simplify the fraction upon creation

    def add(self, other):
        if not isinstance(other, Fraction):
            raise ValueError("The addition is only supported between two Fraction objects.")

        # Find the Least Common Denominator (LCD)
        lcd = self.denominator * other.denominator

        # Adjust numerators based on the LCD
        new_numerator = (self.numerator * other.denominator) + (other.numerator * self.denominator)

        # Create a new Fraction object for the result
        result = Fraction(new_numerator, lcd)
        return result

    def simplify(self):
        gcd = math.gcd(self.numerator, self.denominator)
        self.numerator //= gcd
        self.denominator //= gcd

        # Ensure the denominator is positive
        if self.denominator < 0:
            self.numerator = -self.numerator
            self.denominator = -self.denominator

    def display(self):
        return f"{self.numerator}/{self.denominator}"

def main():
    # Example usage
    f1 = Fraction(3, 4)
    f2 = Fraction(5, 6)

    print(f"Fraction 1: {f1.display()}")
    print(f"Fraction 2: {f2.display()}")

    # Add the two fractions
    result = f1.add(f2)
    print(f"Sum: {result.display()}")

if __name__ == "__main__":
    main()
```
