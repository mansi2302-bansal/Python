## chessboard pattern

```python
matrix = np.zeros((8,8))
matrix
```

```python
len(matrix)   # this is actually the number of rows, or basically the number of lists in a big list
```

```python
for i in range(0,len(matrix)):
  for j in range(0,len(matrix[0])):
    if (i+j)%2 == 0:
      matrix[i][j] = 1
    else:
      continue
```

```python
matrix
```

## Border matrix

```python
m = np.zeros((5,5))
```

```python
m
```

```python
for i in range(0,len(m)):
  for j in range(0,len(m[0])):
    if (i == 0) or (j == 0) or (i == 4) or (j == 4):
      m[i][j] = 1
    else:
      continue
```

```python
m
```
