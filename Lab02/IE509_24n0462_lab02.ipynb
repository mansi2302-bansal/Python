```python
# taking input of the number of rows and columns of the matrix
rows = int(input("enter the number of rows: "))  # specify number of rows and columns
col = int(input("enter the number of cols: "))

# let the matrix name be "A"
A = []

# taking input of the elements of the matrix
for i in range(rows):
  row = []     # taking input row wise
  for j in range(col):
    a = int(input("enter the value: "))
    row.append(a)
  A.append(row)

# initializing the transposed matrix
B = []
for i in range(col):
  row = []
  for j in range(rows):
    a = 0
    row.append(a)
  B.append(row)

# printing matrix A
print("the matrix A is: ")
for i in A:
  print(i)

# transpose of the matrix A
for i in range(rows):
  for j in range(col):
    B[j][i] = A[i][j]

# printing the transposed matrix
print("the transposed matrix B: ")
for i in B:
  print(i)
```

```python
# taking input of the number of rows and columns of the matrix
rows = int(input("enter the number of rows: "))  # specify number of rows and columns
col = int(input("enter the number of cols: "))

# let the matrix name be "A"
A = []

# taking input of the elements of the matrix
for i in range(rows):
  row = []     # taking input row wise
  for j in range(col):
    a = int(input("enter the value: "))
    row.append(a)
  A.append(row)

#Initialising empty arrays and dictionaries to store the sums and results respectively
row_sums = [0]*rows
col_sums= [0]*col

# initialising dictionaries
row_dict = {}
col_dict = {}

# calculating row sums
for i in range(rows):
  for j in range(col):
    row_sums[i] += A[i][j]

# storing row sums in dictionary
print("row sums: ")
for i in range(rows):
  row_dict[i] = row_sums[i]
print(row_dict)

# calculating col sums
for i in range(rows):
  for j in range(col):
    col_sums[j] += A[i][j]

# storing col sums in dictionary
print("Column sums: ")
for i in range(col):
  col_dict[i] = col_sums[i]
print(col_dict)
```

```python
# taking input of the number of rows and columns of the matrix
rows = int(input("enter the number of rows: "))  # specify number of rows and columns
col = int(input("enter the number of cols: "))

# let the matrix name be "A"
A = []

# taking input of the elements of the matrix
for i in range(rows):
  row = []     # taking input row wise
  for j in range(col):
    a = int(input("enter the value: "))
    row.append(a)
  A.append(row)

# forming the flattened list
flattened_list = []
for i in range(rows):
  for j in range(col):
    flattened_list.append(A[i][j])
print(flattened_list)

# frequency count
value_count = [0]*len(flattened_list)
for i in range(len(flattened_list)):
  for j in range(len(flattened_list)):
    if flattened_list[i] == flattened_list[j]:
      value_count[i] = value_count[i] + 1
print(value_count)

# printing frequency in dictionary
freq_dict = {}
for i in range(len(flattened_list)):
  freq_dict[flattened_list[i]] = value_count[i]
print(freq_dict)
```

```python
# taking input of the number of rows and columns of the matrix
rows = int(input("enter the number of rows: "))  # specify number of rows and columns
col = int(input("enter the number of cols: "))

# let the matrix name be "A"
A = []

# taking input of the elements of the matrix
for i in range(rows):
  row = []     # taking input row wise
  for j in range(col):
    a = int(input("enter the value: "))
    row.append(a)
  A.append(row)

# creating the list of unique elements
ele = []
for i in range(rows):
  for j in range(col):
    if A[i][j] not in ele:
      ele.append(A[i][j])
print(ele)

# frequency of elements
row_freq = []
col_freq = []
for i in range(len(ele)):
  row_freq.append([0]*rows);

for i in range(len(ele)):
  col_freq.append([0]*col);

for i in range(len(ele)):
  for j in range(rows):
    for k in range(col):
      if ele[i] == A[j][k]:
        row_freq[i][j] = row_freq[i][j] +1
print(row_freq)

for i in range(len(ele)):
  for j in range(col):
    for k in range(rows):
      if ele[i] == A[k][j]:
        col_freq[i][j] = col_freq[i][j] +1
print(col_freq)

#creating the dictionary
ele_freq = {ele[i] : {"rows": [row_freq[i]], "columns" : [col_freq[i]]} for i in range(len(ele))}
print(ele_freq);
```
