```python
# taking user input
a = float(input("enter coeff of x^2: "))
b = float(input("enter coeff of x: "))
c = float(input("enter constant: "))

#(a)
# defining the function
def quadraticFunction(a,b,c):
  return  f"{a}x^2 + {b}x + {c}"

# calling function
equation = quadraticFunction(a,b,c)
print("equation = ",equation)

#(b)
def derivativeFunction(a,b,c):
  derivative = f"{2*a}x + {b}"
  return derivative

# calling function
derivative = derivativeFunction(a,b,c)
print("derivative = ",derivative)

#(c)
# finding roots
def rootFinder(a,b,c):
  D = (b**2) - (4*a*c)
  if D>=0:
    r1 = (b-(0.5*D))/(2*a)
    r2 = -r1
  else:
    D = abs(D)
    r1 = (b-complex(0,0.5*D))/(2*a)
    r2 = -r1
  return r1,r2

roots = rootFinder(a,b,c)
print("roots of the equation = ",roots)
```

```python
data = {"apple": 70, "banana": 40, "cherry": 20, "date": 100,"fig": 20, "elderberry": 80, "grape": 35, "honeydew": 40, "kiwi": 100, "lemon": 18,
"mango": 50, "nectarine": 22, "orange": 45, "pear": 100, "plum": 33}

def linear_search(data):
# storing the values in a list
  val = list(data.values())

# finding max and min values
  max = val[0]
  for i in range(len(val)):
    if val[i]>max:
      max = val[i]
      max_index = i
  print(f"maximum value is {max}")

  min = val[0]
  for i in range(len(val)):
    if val[i]<min:
      min = val[i]
      min_index = i
  print(f"minimum value is {min}")

  min_key = list(data.keys())[min_index]
  print(f"The fruit with minimum value is: {min_key}")

  max_key = list(data.keys())[max_index]
  print(f"The fruit with max value is: {max_key}")

linear_search(data)
```

```python
# defining function
def binary_search(A,target):
  if len(A) == 0:
    return f"empty parameter"

  for i in range(len(A)):
    start = 0
    end = len(A[0])-1
# the loop keeps on running unless the if condition is followed.
    for j in range(len(A[i])):
      mid = (start+end)//2
      if A[i][mid] == target:
        return f"True, index : ({i},{mid})"
      elif A[i][mid] <= target:
        start = mid + 1
      elif A[i][mid] >= target:
        end = mid -1
      else:
        continue
  return False

# calling function
binary_search([[1,2,5],[7,8,9],[18,19,20]],8)
```

```python
# taking input of n
n = int(input("n: "))

# defining function to find nth term
def fib(n):
  if n<=1:
    return n
  else:
    return f"nth term is {fib(n-1)+fib(n-2)}"

# calling function
fib(n)
```
