```python
# Taking input from the user
year = int(input("enter a year: "))

# if a year is divisible by 4 and not by 100 then it is leap year definitely
if (year%4 == 0 and year%100 != 0):
  print(f"{year} is a leap year.")
elif (year%4 != 0 and year%100 !=0):
  print(f"{year} is not a leap year")

# When a year is divisible by 4 and also by 100 then there comes two conditions to be checked
# If the year is divisible by 400 then it is leap , if not then it is not
elif (year%100 == 0 and year%400 == 0):
  print(f"{year} is a leap year")
elif (year%100 == 0 and year%400 !=0):
  print(f"{year} is not a leap year")

```

```python
# Asking the user if they want to run the program
while True:
  x = input("do you want to run the program yes/no: ")

  # In case of no the program repeats the first while loop of question again and again which makes the addition of this if statement mandatory that terminates
  # the program on giving the input as "no"
  if (x=="no"):
    break

  # if the user enters yes
  while (x  == "yes"):
    n = int(input("enter the hoe many numbers you want to store in the list: "))
    arr = []
    for i in range(n):
      num = int(input("enter the elements: "))
      arr.append(num)      #appends the entered numbers to the array
    print(arr)
    sum = 0       # initializes the sum variable to zero
    for i in arr:
      sum = sum + float(i)     # Iterates over each element in arr and adds each element to previous

    print(f"sum of the elements in the list = {sum}")

    # Average of the numbers (PART 3)
    Avg = sum/(len(arr))
    print(f"average of the numbers = {Avg}")

    # PART 4
    if Avg<50:
      print("Avg is below 50")
    else:
      print("Avg is above 50")
    break
```

```python
# --------DEFINING FUNCTION----------
factors = []
def is_prime(n):
  # A number that has factors only 1 and themselves is prime. So we need to check if the number is divisible
  # by any of the number in the interval (2,n). So we divide the code in two conditions. When n>1 and when it is not.
  if n>1:
    for i in range(2,n):
      if (n%i == 0):
        print(f"{n} is not prime")
        for i in range(2,n+1):
          if (n%i) == 0:
            factors.append(i)
        print(factors)
      else:
        print(f"{n} is prime")
      break   # In order to stop the loop the moment we get the remainder as zero (As we need atleast one number that gives remainder zero).
  else:
    print(f"{n} is prime")
  return


# checking if the input is valid
# While loop will keep on asking for input unless valid input is entered.
while True:
  num = int(input("enter a integer greater than 1: "))

# checking if the input is valid i.e, it is a number or not also it should be greater than 1.
  if (num)>1 == True:
    print("VALID USER INPUT")
    break   # T0 end the loop  as soon as we get valid input.
  else:
    print("-------- user input INVALID ---------")

# Checking if the entered number is prime or not
is_prime(num)

# To find the smallest factor
min_elt = factors[0]
for i in range(len(factors)):
  if factors[i]<min_elt:
    min_elt = factors[i]
print(f"{min_elt} is the smallest factor of {num}")
```
