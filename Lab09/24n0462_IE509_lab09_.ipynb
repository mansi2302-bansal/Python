```python
melted_df = dfr.melt(var_name='Person_ID', value_name='Name')
melted_df
```

```python
#3
melted_df['Name'] = melted_df['Name'].str.title()
melted_df[['First_Name', 'Last_Name']] = melted_df['Name'].str.split(' ',n=1, expand=True)
melted_df = melted_df.drop(columns=['Name'])
melted_df
```

```python
#4
#Sort the DataFrame by Last_Name in alphabetical order and display the
df_sorted = melted_df.sort_values(by='Last_Name', ascending=True)
df_sorted
```

```python
#5
df_sorted['First_Name_Length'] = df_sorted['First_Name'].str.len()
```

```python
#6
df_filtered = df_sorted[df_sorted['Last_Name'].str.startswith('S',na=False)]
```
