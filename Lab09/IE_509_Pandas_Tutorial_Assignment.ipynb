##**3. Modifying Data**

* df['NewColumn'] = ...: Create a new column.

```python
# Adding a new column 'Bonus' which is 10% of Salary
df['Bonus'] = df['Salary'] * 0.10
```

* remove duplicates using the drop_duplicates() function.

```python
# Removing duplicates
df_no_duplicates = df.drop_duplicates()
print(f"Original shape: {df.shape}")
print(f"Shape after removing duplicates: {df_no_duplicates.shape}")
```

```python
 df.drop_duplicates()
```

* df.drop(): Drop columns or rows.

```python
# Dropping the 'Bonus' column
df.drop('Bonus', axis=1, inplace=True)
```

* df.sort_values(): Sort data by column values.

```python
df.sort_values('Salary', ascending=False)
```

##**4) Handling Missing Data**

* df.isnull(): Check for missing values.

```python
df.isnull().sum()
```

* df.fillna(): Fill missing values with a specified value.

```python
# Fill missing Salary values with 0
df['Salary'].fillna(0, inplace=True)
```

* df.dropna(): Drop rows or columns with missing data.

```python
# Drop rows where any value is NaN
df.dropna(inplace=True)
```

##**5) Aggregation and Grouping**

* df.groupby(): Group data by column(s) and apply aggregation.

```python
# Group by Department and get average Salary
df.groupby('Department')['Salary'].mean()
```

* df.agg(): Perform multiple aggregations.

```python
# Perform multiple aggregations on Salary
df.agg({'Salary': ['mean', 'max', 'min']})
```

##**6) Advanced functions**

#### Merging and Joining DataFrames


*   pd.merge(): Merge two DataFrames based on a common column.










```python
# Suppose we have another DataFrame 'dept_info'
dept_info = pd.DataFrame({
    'Department': ['IT', 'HR', 'Finance', 'Marketing'],
    'Manager': ['John Doe', 'Jane Smith', 'Mike Green', 'Nina Jones']
})

# Merge the two DataFrames
merged_df = pd.merge(df, dept_info, on='Department')
merged_df
```

Pivot Tables

*   df.pivot_table(): Create a pivot table to summarize data.

```python
# Create a pivot table showing the average Salary per Department
df.pivot_table(values='Salary', index='Department', aggfunc='mean')
```

##**9. Reshaping Data**
* df.melt(): Convert a DataFrame from wide to long format.

```python
# Reshape data to have Department and Salary as individual rows
melted_df = df.melt(id_vars=['Name'], value_vars=['Department', 'Salary'])
melted_df
```



*   df.pivot(): Pivot the DataFrame based on column values.




```python
# Pivot the data with EmployeeID as columns and values as Salary
pivoted_df = df.pivot(index='EmployeeID', columns='Name', values='Salary')
```

###**10. Time Series Handling**
* pd.to_datetime(): Convert a column to DateTime.

```python
# Convert HireDate to DateTime
df['HireDate'] = pd.to_datetime(df['HireDate'])
df
```

*   df.resample(): Resample time series data to different frequencies.

```python
# Resample by year and get average Salary
df.set_index('HireDate').resample('Y')['Salary'].mean()
```
