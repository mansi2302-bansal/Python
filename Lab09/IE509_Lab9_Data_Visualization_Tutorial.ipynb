### **Data Visualization**

```python
# Let's start by loading the Iris dataset using pandas, and then create some basic visualizations using matplotlib.
# We will begin by importing the necessary libraries and loading the dataset.

import pandas as pd
import matplotlib.pyplot as plt

# Load the Iris dataset
url = "https://archive.ics.uci.edu/ml/machine-learning-databases/iris/iris.data"
columns = ['sepal_length', 'sepal_width', 'petal_length', 'petal_width', 'species']

iris = pd.read_csv(url, header=None, names=columns)

# Display the first few rows of the dataset
print(iris.head())
```

### 1. Line Plot

```python
# Calculate the average sepal length for Iris-setosa at different sepal widths
average_data_setosa = iris[iris['species'] == 'Iris-setosa'].groupby('sepal_width')['sepal_length'].mean()

# Create a figure for the plot
plt.figure(figsize=(10, 6))

# Plot for Iris-setosa
plt.plot(average_data_setosa.index, average_data_setosa, marker='o', linestyle='-', color='red', label='Iris-setosa')

# Add titles and labels
plt.title("Average Sepal Length vs Sepal Width (Iris-setosa)")
plt.xlabel("Sepal Width (cm)")
plt.ylabel("Average Sepal Length (cm)")
plt.legend()
# plt.grid(True)  # Optional: Add a grid for better readability
plt.xticks(average_data_setosa.index)  # Optional: Set x-ticks to match the index
plt.show()
```

Explanation

1. Filtering the Data:

    *   The line `average_data_setosa = iris[iris['species'] == 'Iris-setosa'].groupby('sepal_width')['sepal_length'].mean()` filters the dataset to include only the rows for `Iris-setosa` and calculates the average `sepal length` for different `sepal widths`.

2. Create a Figure:
    *  `plt.figure(figsize=(10, 6))` initializes a new figure for the plot with a specified size of 10 inches wide and 6 inches tall. This provides an appropriate aspect ratio for displaying the plot.

3. Plotting:
    * `plt.plot()` is called to create a line plot:
    * `average_data_setosa.index` provides the x-values (sepal widths).
    * `average_data_setosa` provides the y-values (average sepal lengths).
    * `marker='o'` specifies that circular markers should be placed at each data point.
    * `linestyle='-'` indicates that the data points should be connected with a solid line.
    * `color='red'` sets the line and marker color to red.
    * `label='Iris-setosa'` assigns a label for the plot legend, identifying the species.

4. Adding Titles and Labels:
    * `plt.title()` sets the title of the plot.
    * `plt.xlabel()` and `plt.ylabel()` specify the labels for the x-axis and y-axis, respectively. This helps the viewer understand what the axes represent.

5. Legend and Grid:
    * `plt.legend()` displays the legend on the plot, which helps identify what each line represents.
    * `plt.grid(True)` adds a grid to the background, enhancing the readability of the plot.
    * `plt.xticks(average_data_setosa.index)` sets the x-ticks to match the index of the average data, ensuring the ticks correspond to the sepal widths represented in the data.

6. Display the Plot:
    * `plt.show()` renders the plot to the screen. This is the final step that visualizes all the elements and data added in previous steps.

### 2. Histograms

A histogram helps in understanding the distribution of each feature. For the Iris dataset, we can plot histograms for each feature (sepal length, sepal width, petal length, and petal width).

```python
# Plot histograms iteratively for each feature
for column in iris.columns[:-1]:  # Exclude the 'species' column
    plt.figure(figsize=(6, 4))
    plt.hist(iris[column], bins=20, color='skyblue', edgecolor='black')
    plt.title(f'Histogram of {column}')
    plt.xlabel(column)
    plt.ylabel('Frequency')
    plt.grid(True)
    plt.show()
```

Explanation:


*   for column in `iris.columns[:-1]`: Iterates over all the columns except the last one (species), which is categorical.

*   `plt.hist()`: Creates a histogram for each feature.

*   `bins` parameter determines the number of bins or intervals into which the data will be divided.

### 3. Scatter Plot
A scatter plot helps in visualizing the relationship between two features. We can look at the relationship between sepal length and sepal width for the different species of Iris.

The `plt.scatter()` function in Matplotlib is used to create scatter plots, which are useful for visualizing the relationship between two numerical variables.

```python
# Scatter plot for sepal length vs. sepal width
colors = {'Iris-setosa': 'red', 'Iris-versicolor': 'blue', 'Iris-virginica': 'green'}

plt.figure(figsize=(8, 6))
for species, color in colors.items():
    species_data = iris[iris['species'] == species]
    plt.scatter(species_data['sepal_length'], species_data['sepal_width'], label=species, color=color)

plt.title("Sepal Length vs Sepal Width")
plt.xlabel("Sepal Length (cm)")
plt.ylabel("Sepal Width (cm)")
plt.legend()
plt.show()
```

### 4. Pair Plot

A pair plot (using seaborn) shows scatter plots for all combinations of features and histograms on the diagonal.

```python
import seaborn as sns
```

```python
import warnings
warnings.filterwarnings('ignore')

# Pair plot of the Iris dataset
sns.pairplot(iris, hue="species", diag_kind="hist")
plt.suptitle("Pair Plot of Iris Dataset", y=1.02)
plt.show()
```

Explanation:



1. `import seaborn as sns`: This imports the Seaborn library, which is built on top of Matplotlib and provides a high-level interface for drawing attractive statistical graphics.

2. `sns.pairplot(iris, hue="species", diag_kind="hist")`:
    *   `iris`: The DataFrame containing the Iris dataset.
    *  `hue="species"`: This parameter colors the data points according to the species of the Iris flowers (`Iris setosa, Iris versicolor, Iris virginica`). This allows for easy visual differentiation between the species.
    *   `diag_kind="hist"`: This specifies that the diagonal plots should be histograms. The diagonal plots typically show the distribution of each feature.

3. `plt.suptitle("Pair Plot of Iris Dataset", y=1.02)`: This sets the overall title for the pair plot. The parameter y=1.02 adjusts the vertical position of the title to avoid overlap with the plots.

4. `plt.show()`: This command displays the plot.

### 5. Box Plot
A box plot displays the distribution of a feature and highlights the median, quartiles, and potential outliers.

```python
# Box plot for sepal length grouped by species
plt.figure(figsize=(8, 6))
sns.boxplot(x="species", y="sepal_length", data=iris)
plt.title("Box Plot of Sepal Length by Species")
plt.show()
```

### 6. Heatmap visualization

```python
# Compute the correlation matrix
correlation_matrix = iris.iloc[:, :-1].corr()

# Plot the heatmap for the correlation matrix
plt.figure(figsize=(8, 6))
sns.heatmap(correlation_matrix, annot=True, cmap='coolwarm', linewidths=0.5)

# Set plot title
plt.title('Correlation Matrix of Iris Features')
plt.show()
```

Explanation:

  * `corr()` computes the correlation matrix between the numeric columns.

  *   `sns.heatmap()` creates a heatmap to visualize the correlation values, with the `annot=True` option showing the correlation coefficients directly on the heatmap.
  * The color map `coolwarm` is used to differentiate positive and negative correlations visually.
  * This will help you see which features are positively or negatively correlated with each other.




### 7. Kernel Density Estimates (KDE)

Kernel Density Estimate (KDE) is a non-parametric way to estimate the probability density function of a random variable. In simpler terms, it provides a smoothed curve that represents the distribution of data points over a continuous interval or range

```python
# Set the style of the seaborn plot
sns.set(style='whitegrid')

# Create a figure for the KDE plots
plt.figure(figsize=(12, 8))

# Create KDE plots for each feature colored by species
for feature in ['sepal_length', 'sepal_width', 'petal_length', 'petal_width']:
    plt.subplot(2, 2, ['sepal_length', 'sepal_width', 'petal_length', 'petal_width'].index(feature) + 1)
    sns.kdeplot(data=iris, x=feature, hue='species', common_norm=False, fill=True, alpha=0.5)
    plt.title(f'KDE Plot of {feature}')
    plt.xlabel(feature)
    plt.ylabel('Density')

plt.tight_layout()
plt.show()
```

### 8. Plotting Violin Plot

```python
# Set the style of the seaborn plot
sns.set(style='whitegrid')

# Create a figure for the violin plots
plt.figure(figsize=(12, 8))

# Create violin plots for each feature colored by species
for feature in ['sepal_length', 'sepal_width', 'petal_length', 'petal_width']:
    plt.subplot(2, 2, ['sepal_length', 'sepal_width', 'petal_length', 'petal_width'].index(feature) + 1)
    sns.violinplot(x='species', y=feature, data=iris, inner='quartile')
    plt.title(f'Violin Plot of {feature} by Species')
    plt.xlabel('Species')
    plt.ylabel(feature)

plt.tight_layout()
plt.show()
```
