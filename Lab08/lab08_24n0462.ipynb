```python
import numpy as np
arr = np.array([1,2,3])
arr2 = np.array([4,5,6])
v_norm = np.sqrt(sum(arr2**2))
proj_of_u_on_v = (np.dot(arr, arr2)/v_norm**2)*arr2
print(proj_of_u_on_v)
```
